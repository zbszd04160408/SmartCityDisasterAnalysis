```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
twitter_df = pd.read_csv("./data/cleaned_twitter.csv", index_col = 0)
twitter_df
```

```python
keywords_water = twitter_df.loc[(twitter_df['CleanText'].str.contains("water", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("drink", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("drinking", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("thirst", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("dehydration", case=False, na=False))]
keywords_water
```

```python
keywords_food = twitter_df.loc[(twitter_df['CleanText'].str.contains("food", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("starve", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("hungry", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("milk", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("bread", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("eat", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("foodstuff", case=False, na=False))]
keywords_food
```

```python
keywords_shelter = twitter_df.loc[(twitter_df['CleanText'].str.contains("Shelter", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("house", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("living place", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("sleep", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("rest", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("accommodation", case=False, na=False))]
keywords_shelter
```

```python
keywords_med = twitter_df.loc[(twitter_df['CleanText'].str.contains("Medicine", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("clinic", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("hospital", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("medicine", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("doctor", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("nurse", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("syrup", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("first aid", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("tonic", case=False, na=False))]
keywords_med
```

```python
keywords_electricity = twitter_df.loc[(twitter_df['CleanText'].str.contains("Electricity", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("power", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("electricity", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("light", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("fan", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("energy", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("current", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("charge", case=False, na=False)) |\
                              (twitter_df['CleanText'].str.contains("air conditionor", case=False, na=False))]
keywords_electricity
```

```python
plt.figure(figsize=(10, 8))
plt.bar(x=["Food", "Water", "Medicine", "Electricity", "Shelter"], height=[len(keywords_food), len(keywords_water), len(keywords_med), len(keywords_electricity), len(keywords_shelter)], width=0.3, color = "skyblue")
plt.title("Number of Rweets in Various Categories")
plt.box(False)
plt.gca().yaxis.grid(True)
plt.show()
```

```python
def label (row):
   if "Electricity" in row['CleanText'].lower() or \
         "power" in row['CleanText'].lower() or \
         "electricity" in row['CleanText'].lower() or \
         "light" in row['CleanText'].lower() or \
         "fan" in row['CleanText'].lower() or \
         "energy" in row['CleanText'].lower() or \
         "current" in row['CleanText'].lower() or \
         "charge" in row['CleanText'].lower() or \
         "air conditionor" in row['CleanText'].lower() :
      return 'Electricity'
   elif "clinic" in row['CleanText'].lower() or \
         "hospital" in row['CleanText'].lower() or \
         "medicine" in row['CleanText'].lower() or \
         "doctor" in row['CleanText'].lower() or \
         "nurse" in row['CleanText'].lower() or \
         "syrup" in row['CleanText'].lower() or \
         "first aid" in row['CleanText'].lower() or \
         "tonic" in row['CleanText'].lower():
      return "Medicine"
   elif "shelter" in row['CleanText'].lower() or \
         "house" in row['CleanText'].lower() or \
         "living place" in row['CleanText'].lower() or \
         "sleep" in row['CleanText'].lower() or \
         "rest" in row['CleanText'].lower() or \
         "accommodation" in row['CleanText'].lower():
      return "Shelter"
   elif "food" in row['CleanText'] or \
         "starve" in row['CleanText'] or \
         "hungry" in row['CleanText'] or \
         "milk" in row['CleanText'] or \
         "bread" in row['CleanText'] or \
         "eat" in row['CleanText'] or \
         "foodstuff" in row['CleanText']:
      return "Food"
   elif "water" in row['CleanText'] or\
         "drink" in row['CleanText'] or\
         "drinking" in row['CleanText'] or\
         "thirst" in row['CleanText'] or\
         "dehydration" in row['CleanText']:
      return "Water"
   else:
      return "Other"
   
```

```python
twitter_df = twitter_df[twitter_df['CleanText'].notna()]
twitter_df['Demand'] = twitter_df.apply (lambda row: label(row), axis=1)
```
