```python
import pandas as pd
import numpy as np
import re
```

```python
twitter_df = pd.read_csv("./data/twitter_content.csv", index_col=0)
twitter_df
```

## Split Date and Time

```python
twitter_df['Dates'] = pd.to_datetime(twitter_df['Datetime']).dt.date
twitter_df['Time'] = pd.to_datetime(twitter_df['Datetime']).dt.time
twitter_df
```

## Remove URL, mentions, hashtags, numbers and foreign languages from Tweets

```python
twitter_df['CleanText'] = twitter_df['Text'].replace(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', regex=True)
twitter_df['CleanText'] = twitter_df['CleanText'].replace(r'@[A-Za-z0-9_]+', '', regex=True)
twitter_df['CleanText'] = twitter_df['CleanText'].replace(r'#[A-Za-z0-9_]+', '', regex=True)
twitter_df = twitter_df[twitter_df.CleanText.map(lambda x: x.isascii())]
```

```python
twitter_df.to_csv("./data/cleaned_twitter.csv")
```
